## creating data to view and fit

```python
import numpy as np
import matplotlib.pyplot as plt

# create features
x = np.array([-7.0 , -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

#create labels
y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])
```

```python
y == x + 10
```

```python
# Visualize it
plt.scatter(x , y , color = 'red')
```

# steps in modelling with TensorFlow 

#### 1. **Creating a model** - define the input and output layers, as well as the hidden layers of a deep learning model.
#### 2. **Compiling a model** - define the loss function (in other words, the function which tells our model how wrong it is) and the optimizer (tells our model how to improve the patterns its learning) and evaluation metrics (what we can use to interpret the performance of our model).
#### 3. **Fitting a model** - letting the model try to find patterns between X & y (features and labels). 

```python
# Turn our NumPy to arrays into tensors
x = tf.constant(x)
y = tf.constant(y)

# Reshape x to 2D using tf.reshape
x = tf.reshape(x, (-1, 1))  # Reshape to (n_samples, 1)
x , y
```

```python
# set random seed
tf.random.set_seed(42)

# 1. Create a model using the sequential API
model = tf.keras.Sequential([tf.keras.layers.Dense(1)])

# 2. Compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

# 3. Fit the model
model.fit(x , y , epochs=5)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[17.0]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 17.0: {prediction}")
```

## Improving our model

we can improve our model, by altering the steps we took to create a model.

1. **Creating a model** - here we might add more layres, increase the number of hidden units ( all called neurons) within each of the hidden layers, change the activation function of each layer.
2. **Compiling a model** - here we might change the optimization function or perhaps the Learning rate of the optimization function.
3. **fitting a model** - here we might fit a model for more epochs (leave it training for longer) or on more data (gives the model more examples to learn from)

```python
# 1. create a model
model = tf.keras.Sequential([tf.keras.layers.Dense(1)])

# 2. compile the model 
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

# 3. fit the model
model.fit(x , y, epochs=100)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[17.0]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 17.0: {prediction}")
```

```python
# 1. create a model (Adding layers, increase the number of hidden units and change the activation function)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(1)])

# 2. compile the model 
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

# 3. fit the model
model.fit(x , y, epochs=100)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[17.0]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 17.0: {prediction}")
```

```python
# 1. create a model (Adding layers, increase the number of hidden units and change the activation function)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(1)])

# 2. compile the model (change the optimization function and learning rate)
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=['mae'])

# 3. fit the model
model.fit(x , y, epochs=100)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[17.0]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 17.0: {prediction}")
```

```python
# 1. create a model (Adding layers, increase the number of hidden units and change the activation function)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(100 , activation='relu'),
    tf.keras.layers.Dense(1)])

# 2. compile the model (change the optimization function and learning rate)
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              metrics=['mae'])

# 3. fit the model (with more epochs)
model.fit(x , y, epochs=1000)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[17.0]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 17.0: {prediction}")
```

```python
# Let's have look at how to build a neural network for our data

# Ensure x_train has the correct shape
x_train = np.array(x_train)  # Convert to NumPy array if not already
x_train = x_train.reshape(-1, 1)  # Reshape to (n_samples, n_features)

# 1. Create a model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# 2. Compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=['mae'])

# 3. Fit the model
model.fit(x_train, y_train , epochs=100)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[17.0]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 17.0: {prediction}")
```

```python
# Ensure x_train has the correct shape
x_train = np.array(x_train)  # Convert to NumPy array if not already
x_train = x_train.reshape(-1, 1)  # Reshape to (n_samples, n_features)

# 1. Creat a model with 3 neural network layer
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

# 2. Compile the model 
model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ['mae'])

# 3. Fit the model
model.fit(x_train , y_train , epochs=1000)
```

```python
# Create the input data as a TensorFlow tensor or NumPy array
input_data = tf.constant([[100]], dtype=tf.float32)  # or use np.array([[17.0]])

# Make a prediction
prediction = model.predict(input_data)

print(f"Prediction for input 100: {prediction}")
```
